````python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
X_full = pd.read_csv('./input/iowa-housing/train.csv', index_col='Id')

X_test_full = pd.read_csv('./input/iowa-housing/test.csv', index_col='Id')

# Remove rows with missing target, separate target from predictors
X_full.dropna(axis=0, subset=['SalePrice'], inplace=True) # makes a copy of the data without the rows with missing target values (axis=0 means rows, inplace=True means the operation is performed in place and the original data is modified
y = X_full.SalePrice # y is the target, SalePrice is the column name
X_full.drop(['SalePrice'], axis=1, inplace=True) # drop the SalePrice column from the original data

# To keep things simple, we'll use only numerical predictors
X = X_full.select_dtypes(exclude=['object']) # makes a copy of the data and selects only the columns with numerical data types. Here we are excluding the object data types (which are strings).

X_test = X_test_full.select_dtypes(exclude=['object']) # makes a copy of the data and selects only the columns with numerical data types. Here we are excluding the object data types (which are strings).

# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0) # splits the data into training and validation sets. train_size=0.8 means 80% of the data is used for training, test_size=0.2 means 20% of the data is used for validation, random_state=0 means the split is reproducible. train_test_split returns four objects: the features for the training set, the features for the validation set, the target for the training set, and the target for the validation set.
````

To compare different approaches to dealing with missing values, you'll use the same `score_dataset()` function from the tutorial.  This function reports the [mean absolute error](https://en.wikipedia.org/wiki/Mean_absolute_error) (MAE) from a random forest model.

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
````

#  Cook's Distance

It is useful to know which observations are influential. Cook's distance is a measure of how much a single observation influences the regression line. It is calculated as follows:

$$D_i = \frac{e_i^2}{p\hat{\sigma}^2(h_i)}$$

where $e_i$ is the residual for observation $i$, $p$ is the number of predictors, $\hat{\sigma}^2$ is the estimated variance of the error term, and $h_i$ is the leverage of observation $i$. The leverage of an observation is the distance between the observation and the centroid of the data, divided by the average distance between observations. The centroid of the data is the average of all observations. The average distance between observations is the average of the distances between each observation and every other observation.

The Cook's distance is a measure of how much the regression line changes when a single observation is removed. The Cook's distance for an observation is high when the regression line changes a lot when the observation is removed and vice versa

An outlier is an observation with a high Cook's distance and it is far from the regression line. Any distance over 4/n is considered high. The Cook's distance is normalized by dividing by the number of observations.
